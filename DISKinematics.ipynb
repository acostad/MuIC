```python
import numpy as np
from matplotlib import pyplot as plt
#plt.style.use('tdrstyle.mplstyle')
```

```python
def Q2vsXEp(x,Ep, Elep, Ehad):
    s = 4.0*Elep*Ehad
    #    print("center of mass squared is: ",s)
    tempQ2 = ((1-Ep/Elep)/((1/(s*x))-(1/(4*Elep*Elep))))
    if tempQ2 > 0 :
         y = tempQ2/(s*x)
         if y > 1 :
           print("exceed y=1")
           return s*x
         else : 
           return tempQ2
    else:
         print("negative Q2")
         return 1
```

```python
print("Q2 is: ",Q2vsXEp(0.057,45.0,30.0,820.))
```

```python
def Q2vsXTh(x,TheDeg, Elep, Ehad):
    s = 4.0*Elep*Ehad
    Theta = (TheDeg/180.)*np.pi
    tempQ2 = s/((1/x)+Ehad/Elep*( (1-np.cos(Theta))/(1+np.cos(Theta)) ))
    return tempQ2
```

```python
print("Q2 is: ",Q2vsXTh(0.1,150.,30.0,820.))
```

```python
def Q2vsXEhadp(x,Ep, Elep, Ehad):
    s = 4.0*Elep*Ehad
    #    print("center of mass squared is: ",s)
    #tempQ2 = (((s*x)-4.0*Ep*Elep)/(1.0 - 4*Elep*Elep/(s*x)))
    tempy = Ep/Elep
    if x < Elep/Ehad and tempy > Ehad/Elep :
      tempQ2 = s*x*tempy # approx result for large y
    else :
      tempQ2 = (((s*x)-4.0*Ep*Elep)/(1.0 - 4*Elep*Elep/(s*x)))
    if tempQ2 > 0 :
         y = tempQ2/(s*x)
         if y > 1 :
           print("exceed y=1")
           return s*x
         else : 
           return tempQ2
    else:
         print("negative Q2")
         return 1
#         return -tempQ2
```

```python
print("Q2 is: ",Q2vsXEhadp(0.1,75.0,30.0,820.))
```

```python
def Q2vsXThHad(x,TheDeg, Elep, Ehad):
    s = 4.0*Elep*Ehad
    Theta = (TheDeg/180.)*np.pi
    tempQ2 = s*x/((1) +Elep/(x*Ehad)*( (1+np.cos(Theta))/(1-np.cos(Theta)) ))
    return tempQ2
```

```python
print("Q2 is: ",Q2vsXThHad(0.2,17.2,30.0,820.))
```

```python
nx = 60
xmax = 1.0  # for E'> E
xmin = 4E-6 # For E' < E
xmax0 = Em0/Ehad
xmax1 = Em1/Ehad*0.999
xmax2 = Em2/Ehad*0.999
xmax3 = Em3/Ehad*0.999
xmax4 = Em4/Ehad*0.999
xmax5 = Em5/Ehad*0.999
logxmax0 = np.log(xmax0)
logxmax1 = np.log(xmax1)
logxmax2 = np.log(xmax2)
logxmax3 = np.log(xmax3)
logxmax4 = np.log(xmax4)
logxmax5 = np.log(xmax5)
logxm0 = [0] * nx
logxm1 = [0] * nx
logxm2 = [0] * nx
logxm3 = [0] * nx
logxm4 = [0] * nx
logxm5 = [0] * nx
xm0 = [0] * nx
xm1 = [0] * nx
xm2 = [0] * nx
xm3 = [0] * nx
xm4 = [0] * nx
xm5 = [0] * nx
Q2m0 = [0] * nx
Q2m1 = [0] * nx
Q2m2 = [0] * nx
Q2m3 = [0] * nx
Q2m4 = [0] * nx
Q2m5 = [0] * nx
dlogxm0 = (np.log(xmax0)-np.log(xmin))/nx
dlogxm1 = (np.log(xmax1)-np.log(xmin))/nx
dlogxm2 = (np.log(xmax2)-np.log(xmin))/nx
dlogxm3 = (np.log(xmax3)-np.log(xmin))/nx
dlogxm4 = (np.log(xmax4)-np.log(xmin))/nx
dlogxm5 = (np.log(xmax5)-np.log(xmin))/nx
```

```python
for i in range(nx):
  logxm0[i] = logxmax0 - dlogxm0*i
  logxm1[i] = logxmax1 - dlogxm1*i
  logxm2[i] = logxmax2 - dlogxm2*i
  logxm3[i] = logxmax3 - dlogxm3*i
  logxm4[i] = logxmax4 - dlogxm4*i
  logxm5[i] = logxmax5 - dlogxm5*i
  xm0[i] = np.exp(logxm0[i])
  xm1[i] = np.exp(logxm1[i])
  xm2[i] = np.exp(logxm2[i])
  xm3[i] = np.exp(logxm3[i])
  xm4[i] = np.exp(logxm4[i])
  xm5[i] = np.exp(logxm5[i])
  Q2m0[i] = Q2vsXEhadp(xm0[i],Em0, Elep, Ehad)
  Q2m1[i] = Q2vsXEhadp(xm1[i],Em1, Elep, Ehad)
  Q2m2[i] = Q2vsXEhadp(xm2[i],Em2, Elep, Ehad)
  Q2m3[i] = Q2vsXEhadp(xm3[i],Em3, Elep, Ehad)
  Q2m4[i] = Q2vsXEhadp(xm4[i],Em4, Elep, Ehad)
  Q2m5[i] = Q2vsXEhadp(xm5[i],Em5, Elep, Ehad)
```
