```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('tdrstyle.mplstyle')
```

```python
def EtaLepQ2(Elep, Q2):
    if Q2 > 0 :
      return -np.log(2.0*Elep/np.sqrt(Q2))
    else :
      print("negative Q2")
      return 1
```

```python
print("Elep=1000, Q2=3 => eta= ", EtaLepQ2(1000,100))
```

```python
Elepton = [50., 75., 100., 150., 200., 350., 500., 750., 1000., 1250., 1500.]
```

```python
nbins = 11
```

```python
Etalepton1 = [0]*nbins
Etalepton2 = [0]*nbins
Etalepton3 = [0]*nbins
Etalepton4 = [0]*nbins
```

```python
Q2lepton = [1.0, 3.0, 10., 30.]
```

```python
for i in range(nbins):
  Etalepton1[i] = EtaLepQ2(Elepton[i],Q2lepton[0])
  Etalepton2[i] = EtaLepQ2(Elepton[i],Q2lepton[1])
  Etalepton3[i] = EtaLepQ2(Elepton[i],Q2lepton[2])
  Etalepton4[i] = EtaLepQ2(Elepton[i],Q2lepton[3])
```

```python
plt.plot(Elepton, Etalepton1, 'r')
plt.plot(Elepton, Etalepton2, 'b')
plt.plot(Elepton, Etalepton3, 'g')
plt.plot(Elepton, Etalepton4, 'k')
plt.title('DIS Scattered Lepton Eta vs. Beam Energy')
plt.ylabel('Eta')
plt.xlabel('Incident Beam Energy (GeV)')
plt.figtext(0.52, 0.25, 'Q2 = 1', fontsize=8, color='r')
plt.figtext(0.52, 0.33, 'Q2 = 3', fontsize=8, color='b')
plt.figtext(0.52, 0.41, 'Q2 = 10', fontsize=8, color='g')
plt.figtext(0.52, 0.49, 'Q2 = 30', fontsize=8, color='k')
```
